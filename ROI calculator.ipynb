1. Income                                         
- rental income ~2000                                     
- laundry, storage, misc

2. Expenses                                           
- tax, insurance, utilities, HOA, lawn/snow, vacancy, repairs                                               
- capex (capital expenditures)                               
    -saving up for big things, appliances, roof, etc
- property mgmt, mortgage
- total ~1600

3. Cash Flow
- (income - expenses) = total monthly cash flow 
- ~400

4. Cash on Cash ROI
- down payment ~40k
- closing costs ~3k
- repair/rehab budget ~7k
- misc
- total investment ~50k
- annual cash flow (400x12) = 4800
- annual cash flow 4800 / total investment 50k = 9.6% --> cash on cash roi

```python
class ROICalc:
    print("Welcome! Let's calculate your cash on cash return on investment. ")
    def __init__(self):
        self.total_inc = 0
        self.total_exp = 0
        self.total_inv = 0
        self.annual_cash_flow = 0

    def get_income(self):
        monthly_inc = int(input('Please enter your monthly income '))
        ask = input('Do you have any income from laundry, storage, misc items? y/n ')
        if ask == 'n':
            self.total_inc = monthly_inc
        if ask == 'y':
            other_items = ['laundry', 'storage', 'misc']
            misc_inc = 0
            for item in other_items:
                misc_inc += int(input(f'Please enter your income from {item}: '))
            self.total_inc = monthly_inc + misc_inc
        print(f'Your total monthly income is ${self.total_inc} ')
        
    def get_expenses(self):
        print('='*35)
        print("Let's calculate your expenses. ")
        expenses = [
            'tax', 'insurance', 'utilities', 'HOA', 
            'lawn/snow', 'vacancy', 'repairs', 
            'capital expenditures', 'property management', 'mortgage'
            ]
        for expense in expenses:
            self.total_exp += int(input(f'Please enter your monthly expenses from {expense}: '))
        print(f'Your total expenses are ${self.total_exp}')
    
    def calculate_cash_flow(self):
        total_cash_flow = self.total_inc - self.total_exp
        self.annual_cash_flow = total_cash_flow * 12
        print(f'Your total monthly cash flow is ${total_cash_flow} making your annual cash flow ${self.annual_cash_flow}')
    
    def get_investments(self):
        print('='*35)
        print("Let's calculate your investments. ")
        investments = ['down payment', 'closing costs', 'repair budget', 'miscellaneous']
        for investment in investments:
            self.total_inv += int(input(f'Please enter your investment from {investment}: '))
        print(f'Your total investments are ${self.total_inv} ')
    
    def calculate_cash_roi(self):
        print('='*35)
        self.calculate_cash_flow()
        total_cash_roi = self.annual_cash_flow / self.total_inv
        print(f'Your total cash on cash ROI is {total_cash_roi * 100:.2f}% ')
        

calc = ROICalc()
calc.get_income()
calc.get_expenses()
calc.get_investments()
calc.calculate_cash_roi()
```
